```python
import pandas as pd
```

```python
# reading data set
data_set = pd.read_csv(r'C:\Users\admin\Desktop\data science\assignment\dataset_Facebook.csv', sep = ';')
data_set.head()
```

```python

print(data_set.columns)
```

```python
# divinding the first subset
sub_set_one = data_set[['Post Month', 'Paid']]
```

```python
print("shape of subset is ", sub_set_one.shape)
print(sub_set_one)
```

```python
# divinding the second subset
sub_set_two = data_set[['Page total likes', 'Type', 'Category']]
print("shape of subset is ", sub_set_two.shape)
print(sub_set_two)
```

```python
# merging 
sub_set_three = pd.concat([sub_set_one, sub_set_two ])
print("shape of subset is ", sub_set_three.shape)
sub_set_three.head(10)
```

```python
sub_set_three = sub_set_three.sort_values(by = 'Paid', ascending = True)
```

```python
sub_set_three.transpose()
print("data set after trasnpose is ", sub_set_three.transpose().shape)
```

```python
sub_set_three.head(10)
```

```python
# reshaping
sub_set_four = pd.pivot_table(sub_set_three, index='Paid')
```

```python
print("data set shape is ", sub_set_four.shape)
sub_set_four.head()
```
